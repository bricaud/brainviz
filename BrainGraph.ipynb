## Graph manipulations: Example of the brain graph

Let us load the brain adjacency matrix $A$. We first load the matlab mat file.

```python
import networkx as nx
import scipy as sp
import scipy.io as spio
D = sp.io.loadmat('/Users/benjamin/Documents/MATLAB/Brain/PH0092_test_timeseries.mat')
```

Now the mat file is stored in the dictionary $D$. Let us see what is inside $D$.

```python
D.keys()
```

The adjacency matrices of 5 different graphs, corresponding to different numbers of nodes are stored in $SC$. WE extract the first 3 of them.

```python
Brainall=D['SC']
Brain1=Brainall[0][0]
Brain2=Brainall[0][1]
Brain3=Brainall[0][2]
```

```python
Brain1.shape
```

```python
Brain3.shape
```

We now use specific tools for graphs with networkx

```python
G1=nx.from_numpy_matrix(Brain1)
G2=nx.from_numpy_matrix(Brain2)
G3=nx.from_numpy_matrix(Brain3)
```

```python
nx.draw(G3)
```

```python
nx.draw_spring(G3)
```

```python
nx.write_graphml(G1, "braingraph1")
nx.write_graphml(G3, "braingraph3")
```
